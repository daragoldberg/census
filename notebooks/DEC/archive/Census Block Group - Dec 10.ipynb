```python
import pandas as pd
import json
import requests
import urllib
import os
```

```python
from censusAPI import myAPI
```

```python
url_base = 'https://api.census.gov/data/2010/dec/sf1?'
var = f'P001001,GEO_ID'

stco = {'09':['001','005','009'],\
        '34':['003','013','017','019','021','023','025','027','029',\
              '031','035','037','039','041'],\
        '36':['005','027','047','059','061',\
         '071','079','081','085','087','103','105','111','119']}

nyc = {'36':['005','047','061','081','085']}
```

```python
frames = []
for st,co in nyc.items():
    for i in co:
        url = f'{url_base}get={var}&for=block%20group:*&in=state:{st}%20county:{i}&key={myAPI}'
        resp = requests.request('GET', url).content
        df = pd.DataFrame(json.loads(resp)[1:])
        df.columns = json.loads(resp)[0]
        frames.append(df)
```

```python
dff = pd.concat(frames)
dff.head()
```

```python
dff['cbg'] = dff['GEO_ID'].str[-12:]
dff['stco'] = dff['state']+dff['county']
dff.head()
```

```python
dff = dff.drop(columns=['GEO_ID','tract','block group']).rename(columns={'P001001':'Pop_10E'})
dff.head()
```
